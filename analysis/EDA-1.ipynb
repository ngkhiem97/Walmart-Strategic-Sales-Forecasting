```python
import pandas as pd
```

# Dataset: Fuel Price

```python
fuel_data = pd.read_csv('../data/raw/PET_PRI_GND_DCUS_NUS_W.csv')
fuel_data.head()
```

```python
fuel_data[fuel_data.columns].plot(figsize=(20, 10))
```

```python
fuel_data["A1"].plot(figsize=(20, 10))
```

```python
fuel_data['month'] = pd.to_datetime(fuel_data['Date']).dt.month
fuel_data['year'] = pd.to_datetime(fuel_data['Date']).dt.year
fuel_data.groupby(['year','month'],as_index=False).mean()["A1"].plot(figsize=(20, 10))
```
