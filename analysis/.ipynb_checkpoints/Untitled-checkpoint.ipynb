```python
import pandas as pd
import numpy as np

calendar_data = pd.read_csv('../data/processed/calendar_processed-May-13-2022.csv')
calendar_data["date"] = pd.to_datetime(calendar_data["date"])
total_sales = calendar_data.set_index("date")['total_sales']
total_sales.plot(figsize=(20, 10))
```

```python
total_sales_diff = pd.DataFrame(np.diff(total_sales))
total_sales_diff.plot(figsize=(20, 10))
```

```python
rolling_deviations = pd.Series(dtype=float, index = total_sales.index)
```

```python
for date in rolling_deviations.index:
    #get the window ending at this data point
    window = total_sales.loc[:date]
    
    #get the deviation within this window
    rolling_deviations.loc[date] = window.std()
```

```python
diff_rolling_deviations = rolling_deviations.diff()
diff_rolling_deviations = diff_rolling_deviations.dropna()
```

```python
import matplotlib.pyplot as plt
from datetime import datetime

plt.figure(figsize=(10,4))
plt.plot(diff_rolling_deviations)
plt.title('Deviation Differences', fontsize=20)
plt.ylabel('Sales', fontsize=16)
```

```python
month_deviations = total_sales.groupby(lambda d: d.month).std()
```

```python
plt.figure(figsize=(10,4))
plt.plot(month_deviations)
plt.title('Deviation by Month', fontsize=20)
plt.ylabel('Sales', fontsize=16)
```

```python
from datetime import timedelta
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings("ignore")

my_order = (0,1,0)
my_seasonal_order = (1, 0, 1, 12)

train_end = datetime(2015,6,19)
test_end = datetime(2016,6,19)

test_data = total_sales[train_end + timedelta(days=1):test_end]

rolling_predictions = test_data.copy()
for train_end in test_data.index:
    train_data = total_sales[:train_end-timedelta(days=1)]
    model = SARIMAX(train_data, order=my_order, seasonal_order=my_seasonal_order)
    model_fit = model.fit(disp=False)

    pred = model_fit.forecast()
    rolling_predictions[train_end] = pred
```

```python
   
rolling_residuals = test_data - rolling_predictions
plt.figure(figsize=(10,4))
plt.plot(rolling_residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Rolling Forecast Residuals from SARIMA Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
```

```python
start_date = datetime(2011,1,29)
end_date = datetime(2016,6,19)

plt.figure(figsize=(10,4))

plt.plot(total_sales)
plt.plot(rolling_predictions)

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('Catfish Sales in 1000s of Pounds', fontsize=20)
plt.ylabel('Production', fontsize=16)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```
